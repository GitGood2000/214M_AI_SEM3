### Задание 3: Попробуйте превратить "starry_night" в "starry_day" путем увеличения яркости изображения и смены темно-синих тонов на светло-желтые. 
### Как это сделать? См. подсказку по смешиванию RGB цветов в файле color_blending.gif. 
#### Замечание: при проведении экспериментов с матрицами каналов, обратите внимание на dtype этих матриц (стандартный тип uint8 не позволит вам выйти за границы диапазона 0 .. 255, и поэтому может быть причиной появления не тех результатов, которые вы ожидаете, поэтому перед проведением операций целесообразно этот тип сменить на более подходящий).

```python
import numpy as np
img_bgr = cv.imread("images/starry_night.jpg")

# Конвертируем изображение из BGR в RGB
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

```python
img_mod = img_rgb
blue_channel = img_mod[:, :, 2]
green_channel = img_mod[:, :, 1]
red_channel = img_mod[:, :, 0]
img_mod[:, :, 1] = red_channel
img_mod[:, :, 0] = green_channel
img_mod[:, :, 2] = blue_channel/1.8
cv.convertScaleAbs(img_rgb, img_mod, 1.8, 10)
plt.imshow(img_mod)
```
