### Проецируем прогресс обучения

```python
import matplotlib.pyplot as plt
from IPython.display import HTML
from matplotlib import animation
import torch
```

```python
imgs = []
for i in range(500):
  if i % 10 == 0:
    imgs.append(torch.tensor(plt.imread(f'/content/drive/MyDrive/Colab Notebooks/img/train_{i}.png')))
```

```python
fig = plt.figure()

# Clamp to avoid complains
im = plt.imshow(imgs[0].clamp(0.0, 1.0), animated=True)

def update_fig(i):
    im.set_array(imgs[i].clamp(0.0, 1.0))
    return im,
anim = animation.FuncAnimation(fig, update_fig,
                               frames=len(imgs), interval=50, blit=True)
HTML(anim.to_jshtml())
```
